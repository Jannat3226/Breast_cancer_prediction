```python
#label Encoading
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df[['diagnosis']]=df[['diagnosis']].apply(LabelEncoder().fit_transform)
```

```python
# Get the count of malignant<M> and Benign<B> cells
df['diagnosis'].value_counts()
```

```python
#one-hot encoding
df=pd.get_dummies(df)
df.head(20)
```

```python
# visualize the correlation
plt.figure(figsize=(10,10))
sns.heatmap(df.iloc[:,1:10].corr(),annot=True,fmt=".0%")
```

```python
#removing outlairs

Q1 = df.quantile(0.20)
Q3 = df.quantile(0.80)
IQR = Q3 - Q1

df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# split the dataset into dependent(X) and Independent(Y) datasets
X=df.iloc[:,2:31].values
Y=df.iloc[:,1].values
```

# Splitting the data into training data & Testing data

```python
# spliting the data into trainning and test dateset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
X_train
```

```python
Y_train
```

# training the Logistic Regression model using Training data

```python
model.fit(X_train, Y_train)
```

```python
model.predict(X_test)
```

```python
Y_predict=model.predict(X_test)
print("Accuracy",(model.score(X_test,Y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test,Y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_predict))
```

```python
#cross validation
from sklearn.model_selection import cross_val_score
crv= cross_val_score(model,X,Y,cv=5)
crv
```
